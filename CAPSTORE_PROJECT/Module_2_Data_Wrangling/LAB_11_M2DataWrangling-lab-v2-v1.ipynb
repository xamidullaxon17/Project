### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
df['Country'].value_counts()
```

```python
# or 
df['Country'].unique()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
mapped_country = {
    'United States of America': 'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'UK',
    'Viet Nam': 'Vietnam',
    'Russian Federation': 'Russia',
    'Iran, Islamic Republic of...': 'Iran',
    'Bolivarian Republic of...': 'Venezuela',
    'Syrian Arab Republic': 'Syria',
    'Czech Republic': 'Czechia',
    'Republic of Korea': 'South Korea',
    "People's Republic of China": 'China',
    'Central African Republic': 'CAR',
    'Democratic Republic of the Congo': 'DRC',
    'Federated States of...': 'FSM',
    'Republic of North Macedonia': 'North Macedonia',
    'United Republic of Tanzania': 'Tanzania',
    'United Arab Emirates': 'UAE',
    'Brunei Darussalam': 'Brunei'
}

df['Country'] = df['Country'].replace(mapped_country)
df['Country'].value_counts
```

```python
df['EdLevel'].unique()
```

```python
edlevel_mapping = {
    'Primary/elementary school': 'Primary',
    'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)': 'Secondary',
    'Some college/university study without earning a degree': 'Some College',
    'Associate degree (A.A., A.S., etc.)': 'Associate',
    'Bachelor’s degree (B.A., B.S., B.Eng., etc.)': 'Bachelor',
    'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)': 'Master',
    'Professional degree (JD, MD, Ph.D, Ed.D, etc.)': 'Professional',
    'Something else': 'Other'
}
df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)
df['EdLevel'].value_counts()
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
# 1. One-hot encoding
employment_encoded = pd.get_dummies(df['Employment'], prefix='Employment') # bu - 
# df['Employment'] dagi unique qiymatlar uchun yangi ustunlar yaratadi va column nomlarini 'Employment_' bilan boshlaydi 

# 2. Asosiy dataframe bilan birlashtirish
df = pd.concat([df, employment_encoded], axis=1)

# df = df.drop('Employment', axis=1)
df
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
missings = df.isnull().sum()
missings.sort_values(ascending=False).head()
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean(), inplace=True)
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
df['RemoteWork'].fillna(df['RemoteWork'].mode()[0], inplace=True)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
df['ConvertedCompYearly_MinMax']
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
# Log-transform (0 yoki manfiy qiymatlar bo'lmasligi kerak)
df['ConvertedCompYearly_Log'] = np.log(df['ConvertedCompYearly'] + 1)  # +1, 0 qiymatlarni oldini olish uchun
df['ConvertedCompYearly_Log']
# Izohlar:
# np.log() tabiiy logarifmni oladi (base e).
# +1 qo‘shish 0 qiymatlar uchun log(0) xatosini oldini oladi.
# Log-transform skewed (ya’ni o‘ngga cho‘zilgan) ma’lumotlarni ko‘proq normal (Gaussian) taqsimotga yaqinlashtiradi.
# Bu usul ko‘pincha kompensatsiya (maosh) kabi juda katta qiymatlar bo‘lgan ustunlar uchun ishlatiladi.
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df['YearsCodePro'].value_counts()
```

```python
df['YearsCodePro'].fillna(0, inplace=True)
```

```python
def get_level(year):
    if pd.isna(year):
        return 'Unknown'
    if year == 'Less than 1 year':
        return 'Junior'
    if year == 'More than 50 years':
        return 'Expert'
    try:
        year = int(year)
    except ValueError:
        return 'Unknown'
    
    if year < 2:
        return 'Junior'
    elif 2 <= year < 5:
        return 'Mid-level'
    elif 5 <= year < 10:
        return 'Senior'
    else:
        return 'Expert'

    
df['ExperienceLevel'] = df['YearsCodePro'].apply(get_level)
df['ExperienceLevel'].value_counts()
```
