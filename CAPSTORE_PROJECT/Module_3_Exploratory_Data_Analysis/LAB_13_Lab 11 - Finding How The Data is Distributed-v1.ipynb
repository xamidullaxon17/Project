### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
worked = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()
worked
```

```python
want = df['LanguageWantToWorkWith'].dropna().str.split(';').explode().reset_index()
want
```

```python
worked_counts = worked.value_counts().head(10) 
worked_counts
```

```python
want_counts = want.value_counts().head(10)
want_counts
```

```python
plt.figure(figsize=(12,6))
worked_counts.plot(kind='bar', alpha=0.7, label='Worked With')
want_counts.plot(kind='bar', alpha=0.7, label='Want To Work With', color='orange')
plt.ylabel("Number of Respondents")
plt.title("Top Programming Languages: Worked With vs Want To Work With")
plt.legend()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
remote_region = pd.crosstab(df['Country'], df['RemoteWork'])
remote_region
```

```python
top_countries = df['Country'].value_counts().head(10).index
remote_region_top = remote_region.loc[top_countries]
remote_region_top
```

```python
plt.figure(figsize=(8,5))
remote_region_top.plot(kind='bar')
plt.title("Remote Work Preference by Top 10 Countries")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(remote_region_top, annot=True, fmt='d', cmap='YlGnBu')
plt.title("Remote Work Heatmap by Country")
plt.ylabel("Country")
plt.xlabel("Remote Work Status")
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
df['YearsCodePro'].value_counts()
```

```python
def get_exp(exp):
    if exp == 'Less than 1 year':
        return 0.5
    elif exp == 'More than 50 years':
        return 51
    
    try:
        return float(exp)
    except:
        return None
df['YearsCodePro'] = df['YearsCodePro'].apply(get_exp)
```

```python
corr = df[['JobSat', 'YearsCodePro']].corr()
corr
```

```python
from scipy.stats import pearsonr, spearmanr
df_corr = df[['JobSat', 'YearsCodePro']].dropna()
pearson_corr, pearson_p = pearsonr(df_corr['JobSat'], df_corr['YearsCodePro'])
spearman_corr, spearman_p = spearmanr(df_corr['JobSat'], df_corr['YearsCodePro'])
print(f"Pearson correlation: {pearson_corr:.3f}, p-value: {pearson_p:.3f}")
print(f"Spearman correlation: {spearman_corr:.3f}, p-value: {spearman_p:.3f}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='YearsCodePro', y='JobSat', data=df_corr, alpha=0.5)
plt.xlabel("Years of Professional Coding Experience")
plt.ylabel("Job Satisfaction (numeric)")
plt.title("Job Satisfaction vs Years of Experience")
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
employment_edu = pd.crosstab(df['EdLevel'], df['Employment'])
employment_edu
```

```python
employment_edu.plot(kind='bar', stacked=True, figsize=(12,6))
plt.ylabel("Number of Respondents")
plt.xlabel("Education Level")
plt.title("Employment Status by Education Level")
plt.legend(title="Employment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
